```python
def scrape(driver):
    df = pd.DataFrame(columns=["Job Title","Location","Company Name","Salary Estimate","Job Description","Rating",
                               "Days Posted", "Job Type","Experience Level","Skills","Time Collected"])
    
    driver.implicitly_wait(random.randint(1,4))

    soup = BeautifulSoup(driver.page_source,'html.parser')

    indeed_containers = soup.find_all('div', attrs={'class': 'jobsearch-SerpJobCard'})

    for container in indeed_containers:

        time_collected = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
#         print(container.text)

        try:
            title = container.find("a",class_="jobtitle").text.replace("\n","").strip()
        except:
            title = None

        try:
            location = container.find(class_="location").text
        except:
            location = None

        try:
            company = container.find(class_="company").text.replace("\n","").strip()
        except:
            company = None

        try:
            salary = container.find(class_="salary").text.replace("\n","").strip()
        except:
            salary = None

#         try:
#             sponsored = container.find(class_="sponsoredGray").text
#             sponsored = "Sponsored"
#         except:
#             sponsored = "Organic"

        try:
            job_desc = container.find(class_='summary').text
        except:
            job_desc = None

        try:
            rating = container.find(class_='ratingsContent').text.replace("\n","")
        except:
            rating = None

        try:
            days_posted = container.find(class_='date').text
        except:
            days_posted = None

        try:
            jobs_type = container.find(class_='jobMetadataHeader-itemWithIcon-icon jobMetadataHeader-itemWithIcon-icon-jobs').findNext('div',{'class':'class_value'})
        except:
            jobs_type = None


        try:
            experience = container.find(class_='jobMetadataHeader-itemWithIcon-icon jobMetadataHeader-itemWithIcon-icon-resume').text       
        except:
            experience = None

        try:
            skills = []
#             soup.find_all('div', attrs={'class': 'jobsearch-SerpJobCard'})
            for skill in container.find_all('button', attrs={'class':'skillLinkSerp'}):
                skills.append(skill.text)
            skills = ', '.join(skills)
#                 print('Yes')
        except:
            skills = None

        df = df.append({"Company Name":company, "Days Posted": days_posted, "Experience Level" : experience, 
                        "Job Description":job_desc, 'Job Title':title, "Job Type" : jobs_type, 'Location':location, 
                        "Rating" : rating, "Salary Estimate":salary, "Skills" : skills, "Time Collected": time_collected},
                       ignore_index=True)
        print(df.shape)
        
    return df
```

```python
# put the location of your webdriver here
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(r"C:\Users\User\Desktop\Fresh\Vertex\Misc\chromedriver\chromedriver")

for j in list(vertex_dict['Location_Indeed'].dropna()):
    df = pd.DataFrame(columns=['Company Name', 'Days Posted', 'Experience Level', 'Job Description', 'Job Title',
                               'Job Type', 'Location', 'Rating', 'Salary Estimate', 'Skills', 'Time Collected'])
    
    for i in range(0,1001,10):
        driver.get(str(j)+'&start='+str(i))
        
        temp_df = scrape(driver)
        df = pd.concat([df, temp_df])
                
        if i == 1000:
            for k in range(0, 1001, 10):
                driver.get(str(j)+'&filter=0&start='+str(k))
                
                temp_df = scrape(driver)
                df = pd.concat([df, temp_df])
                
        try:
            if os.path.isfile(r'C:\Users\User\Desktop\Fresh\Vertex\indeed_full_jobs.csv'):
                df.to_csv(r'C:\Users\User\Desktop\Fresh\Vertex\indeed_full_jobs.csv', mode='a', header=False, encoding='utf-8-sig')
                print('Yes')
            else:
                df.to_csv(r'C:\Users\User\Desktop\Fresh\Vertex\indeed_full_jobs.csv', encoding='utf-8-sig')
        except:
            print('Pass')
            pass

    print(j, df.shape)
```

```python
def get_number(string):
    if re.findall(r'\d+', string):
        return re.findall(r'\d+', string)[0]
    
def get_string(string):
    return ''.join([i for i in string if not i.isdigit()])
```
